#### 1.1
Write a function *count_bigrams_in_file* that takes a filename as its only argument.

####  <font color = red>Construct bigram string slicing</font>

```python
from collections import Counter
import string
import re
def count_bigrams_in_file(file_name):
    try:
        if not isinstance(file_name,str):
            raise TypeError("input file_name as a string")
        #read txt into a string object
        f = open('./' + file_name)
        raw = f.read().lower().strip()
        raw = raw.translate(str.maketrans('','',string.punctuation))
        raw = re.sub('[^a-zA-Z0-9]'," ", raw,flags=re.VERBOSE)
        #filter out the extra white spaces
        clean_s = re.sub(r"\s+", " ",raw)
    
        #bigram list
        bigram_list = tuple(zip(clean_s.split(" ")[:-1], \
                   clean_s.split(" ")[1:]))
        #input into the dictionary
        bigram_dic = dict(Counter(bigram_list))
        #check for empty bigrams components
        if True in [len(j)==0 for i in bigram_dic.keys() for j in i]:
            raise ValueError('txt file not cleaned properly')
        return bigram_dic
    except IOError:
        print("file cannot be opened:", file_name)
```

#### 1.3:  
write a function collocations that takes a filename as its only argument and returns a dictionary

```python
from collections import defaultdict
import string
import re 
def remove_dup(l):
    """
    l: list to be uniqueify
    remove duplicates while mantaining order of the list
    """
    seen = set()
    return [x for x in l if not (x in seen or seen.add(x))]
###################################################################
def collocations(file_name):
    try:
        if not isinstance(file_name,str):
            raise TypeError("input file_name as a string")
        
        #read txt into a string object
        f = open('./' + file_name)
        raw = f.read().lower().strip()
        raw = raw.translate(str.maketrans('','',string.punctuation))
        raw = re.sub('[^a-zA-Z0-9]'," ", raw,flags=re.VERBOSE)
        #filter out the extra white spaces
        clean_s = re.sub(r"\s+", " ",raw)
    
        #bigram list/list with tuples of bigrams
        bigram_list = list(zip(clean_s.split(" ")[:-1], \
                   clean_s.split(" ")[1:]))
    
        #reverse the bigram list for collocations matching reversely
        bigram_list_reverse = [(tup[1],tup[0]) for tup in bigram_list]
    
        #remove duplicates from the bigram_list to make efficient
        #while maintaining order for the ease of grading
        bi_list = remove_dup(bigram_list)
        bi_list_rev = remove_dup(bigram_list_reverse)
    
        #initiate default dictionary in set to ensure uniqness
        dd = defaultdict(set)
        #traverse through all the concanated list of bigrams
        for k, v in bi_list + bi_list_rev:
            dd[k].add(v)
        collocate_dict = dict((k, v) for k, v in dd.items())
        return collocate_dict
    except IOError:
        print("file cannot be opened:", file_name)
```

#### 2.1 & 2.4 & 2.5 & 2.6 & 2.7:
Implement the necessary operator(s) ,**<font color = green>(equality, less than, less
or equal to, greater than, greater or equal to, dot, mul, norm, __str__, __len__)</font>** of Vector objects.

```python
### define a vector class:::(list implementation/not tuple)
class Vector(object):
    #Constructor
    def __init__(self,dimension = 0,entry = None):
        ''' Constructor for Vector dimension should be 
            a non-negative int entries should be ints or floats
            set the default arguments for dimension to be 0,
            In that when Vector() is called, should return default empty vector
            
        '''
        self.dim, self.val = dimension,entry
        
    #decorators for getting dim    
    @property
    def dim(self):
        '''Setter for dimension'''
        return self.__dim
    #decoratos for getting val
    @property
    def val(self):
        '''Setter for entries'''
        return self.__val
    
    #decorators for setting dim
    @dim.setter
    def dim(self,value):
        '''Setter for dimension: includes nessacary error checking '''
        #default error checking
        if value < 0 :
            raise ValueError("dimension should be a non-negtive int")
        #check dimension type
        elif isinstance(value,int) != True:
            raise TypeError("dimension input type can only be int")
        else:
            self.__dim = value
            
    #decorators for setting val
    @val.setter
    def val(self,value):
        '''Setter for entries: includes nessacary error checking '''
        #if entries not enter, entries set to be zero list with matched dimension
        if value == None:
            self.__val = [0]*self.__dim
        #check entry type 
        elif not (isinstance(value, list)):
            raise TypeError("entry type can only be list")
        #check elements within entry type
        elif False in set([type(i) in(float,int)for i in value]):
            raise TypeError("entries element should only be type int or float")
        #check matching dimension
        elif len(value) != self.__dim:
            raise ValueError("dimension should match to the length of the entry")
        else:
            self.__val = value
            
    ##########operator methods###################################################  
    
    #Methods for operator "=="        
    def __eq__(self, other):
        '''Returns whether two vectors are equal'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else: 
            return self.__val == other.__val
        
    #Methods for operator "!="
    def __ne__(self, other):
        '''Returns whether two vectors are not equal'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return self.__val != other.__val
        
    #Methods for operator "<"
    def __lt__(self,other):
        '''Returns whether left < right Vector'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return self.__val < other.__val
        
    #Methods for operator "<="
    def __le__(self,other):
        '''Returns whether left <= right Vector'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return self.__val <= other.__val
        
    #Methods for operator ">"
    def __gt__(self,other):
        '''Returns whether left > right Vector'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return self.__val > other.__val
        
    #Methods for operator ">="
    def __ge__(self,other):
        '''Returns whether left >= right Vector'''
        if type(other) != Vector:
            print("comparing with non-Vector type")
            return False
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return self.__val >= other.__val
        
    #Methods for operator "*" in order like Vec*2 or Vec1*Vec2
    def __mul__(self, other):
        """ Returns the dot product of self and other if multiplied
            by another Vector.  If multiplied by an int or float,
            multiplies each component by other.
        """
        if type(other) == type(self) == Vector: 
            entry = [a * b for a, b in zip(self.__val, other.__val)]
        elif isinstance(other, (int, float)):
            entry = [num * other for num in self.__val]
        else:
            raise TypeError("multiplication applies only to scalar and Vector")
        return Vector(self.__dim, entry)  
    #Methods for operator "*" for 2*Vec
    def __rmul__(self, other):
        """ Called if 4*self for instance """
        return self.__mul__(other)
    ###########other Methods######################################################
    
    #Methods for dot product
    def dot(self,other):
        '''Returns dot product of two Vectors'''
        if type(other) != Vector:
            raise TypeError("dot product with non-Vector type")
        elif self.__dim != other.__dim:
            raise ValueError("comparing two Vectors of different dimesions")
        else:
            return sum(a * b for a, b in zip(self.__val, other.__val))
        
    #Methods for vector norm
    def norm(self,p):
        '''returns the vector norm of object'''
        if not isinstance(p,(int,float)):
            raise TypeError("p should be an int or float")
        elif p < 0:
            raise ValueError("p should be a positive int or float")
        elif p == 0:
            return len([v for v in self.__val if v!=0])
        elif p > 0 and p < float('Infinity'):
            return sum([abs(v)**p for v in self.__val])**(1.0/p)
        else:
            return max(self.__val)
        
    #Methods for "len(Vector())"
    def __len__(self):
        '''returns a length of the entries of the object'''
        return self.__dim
    def __str__(self):
        '''returns a string representation of the object'''
        return "a "+str(self.__dim)+" dimension Vector with entries of "\
                + str(self.__val)
```
